### **Pipeline**

Pipeline merupakan salah satu fungsi basic di Transformers. Fungsi ini menghubungkan sebuah model dengan preprocessing dan postprocessing step-nya yang dapat kita gunakan untuk menginput teks dan mendapatkan jawaban yang jelas. 

Ketika kita menginput teks ke Pipeline, proses yang terjadi adalah:
1. Teks akan dipraproses sesuai format yang dipahami model.
2. Teks yang dipraproses akan di-passing ke model.
3. Hasil prediksi yang dihasilkan model akan di-postprocessing supaya kita dapat memahaminya.

Dengan fungsi ini, kita dapat langsung menemukan hasil prediksi hanya dengan sekali memasukkan input teks!

Kita dapat menggunakan beberapa jenis pipeline yang disediakan library Transformers, sebagai berikut:
*   feature-extraction (get the vector representation of a text)
*   fill-mask
*   ner (named entity recognition)
*   question-answering
*   sentiment-analysis
*   summarization
*   text-generation
*   translation
*   zero-shot-classification

```python
from transformers import pipeline
```

#### 1. Sentiment Analysis

Seperti namanya, pada pipeline ini kita dapat memperoleh hasil sentimen dari sebuah kata, apakah kata tersebut bernilai positif atau negatif beserta skornya.

```python
classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for a HuggingFace course my whole life.")
```

```python
classifier(["Excited to learn Transformers!", "Happy new year!"])
```

#### 2. Zero Shot Classification

Zero shot classification digunakan ketika kita ingin mengklasifikasi sebuah kalimat yang tidak dilabeli. Kita dapat menentukan sendiri label yang kita inginkan, misalnya seperti positif dan negatif seperti sentimen analisis, atau label lainnya.

Selain menggunakan model default, kita juga bisa mencoba untuk menggunakan model-model lain yang ada di https://huggingface.co/models. Model yang digunakan di bawah ini adalah model dari Facebook https://huggingface.co/facebook/bart-large-mnli.

```python
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
sequence_to_classify = "I learn Transformers today"
candidate_labels = ['education', 'fashion', 'politics']
classifier(sequence_to_classify, candidate_labels)
```

Pipeline ini dinamai zero-shot karena kita tidak perlu fine-tuning model ke data kita untuk menggunakannya dan dapat langsung mereturn skor probabilitas untuk setiap list label yang kita pakai.

#### 3. Text Generation

Text generation adalah pipeline yang digunakan untuk meng-generate teks. Kita dapat menginput sebuah teks dan model akan menghasilkan lanjutan (auto-complete) teks tersebut. Model ini meng-generate teks secara acak, jadi hasil yang didapatkan kadang berbeda-beda.

Kita dapat mengatur berapa banyak sequence yang di-generate dengan argumen `num_return_sequences` dan maksimal panjang kata yang di-generate dengan argumen `max_length`.

```python
generator = pipeline("text-generation")
generator(
    "I am studying computer science",
    max_length=50,
    num_return_sequences=2,
)
```

Mari kita coba menggunakan model Bahasa Indonesia dari https://huggingface.co/cahya/gpt2-small-indonesian-522M

```python
generator = pipeline("text-generation", model='cahya/gpt2-small-indonesian-522M')
generator(
    "Hari ini aku pergi ke",
    max_length=30, 
    num_return_sequences=5, 
    num_beams=10
)
```

#### 4. Mask filling

```python
unmasker = pipeline('fill-mask')
unmasker("I am going to <mask> today", top_k=3)
```

#### 5. Named Entity Recognition (NER)

```python
ner = pipeline('ner', grouped_entities=True)
ner("My name is Zahra. I am an undergraduate student at ITS in Surabaya")
```

#### 6. Question Answering

```python
answerer = pipeline('question-answering')
answerer(question="Where do Zahra study?", context="Zahra is an undergraduate student at ITS in Surabaya")
```

#### 7. Summarization

```python
summarizer = pipeline('summarization')
summarizer(
    """
    NLP is a field of linguistics and machine learning focused on understanding everything related to human language. The aim of NLP tasks is not only to understand single words individually, but to be able to understand the context of those words.

    The following is a list of common NLP tasks, with some examples of each:

    Classifying whole sentences: Getting the sentiment of a review, detecting if an email is spam, determining if a sentence is grammatically correct or whether two sentences are logically related or not
    Classifying each word in a sentence: Identifying the grammatical components of a sentence (noun, verb, adjective), or the named entities (person, location, organization)
    Generating text content: Completing a prompt with auto-generated text, filling in the blanks in a text with masked words
    Extracting an answer from a text: Given a question and a context, extracting the answer to the question based on the information provided in the context
    Generating a new sentence from an input text: Translating a text into another language, summarizing a text
    NLP isn’t limited to written text though. It also tackles complex challenges in speech recognition and computer vision, such as generating a transcript of an audio sample or a description of an image.
    """
)
```
